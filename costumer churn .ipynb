```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create a crosstab with normalized (percentage) values
ct = pd.crosstab(df["SeniorCitizen"], df["Churn"], normalize='index') * 100

# Step 2: Plot stacked bar chart
ax = ct.plot(kind='bar', stacked=True, colormap='Set2')

# Step 3: Add percentage labels
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%')

# Step 4: Customize chart
plt.title("Churn Percentage by Senior Citizen Status")
plt.xlabel("Senior Citizen (0 = No, 1 = Yes)")
plt.ylabel("Percentage")
plt.legend(title="Churn")
plt.xticks(rotation=0)
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

**more senior citizens are churning out as compare to others

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Calculate percentage distribution of 'Churn' within each 'Contract' type
ct = pd.crosstab(df["Contract"], df["Churn"], normalize='index') * 100

# Step 2: Plot stacked bar chart
ax = ct.plot(kind='bar', stacked=True, colormap='Set2', figsize=(8, 6))

# Step 3: Add percentage labels
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='center')

# Step 4: Customize chart
plt.title("Churn Percentage by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Percentage")
plt.ylim(0, 100)
plt.legend(title="Churn")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python


# List of service columns
service_columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Filter only churned customers
churned_df = df[df['Churn'] == 'Yes']

# Prepare a dataframe to hold all counts
stacked_data = pd.DataFrame()

for col in service_columns:
    counts = churned_df[col].value_counts()
    stacked_data[col] = counts

# Transpose and fill NaNs with 0
stacked_data = stacked_data.fillna(0).T

# Plotting
ax = stacked_data.plot(kind='bar', stacked=True, figsize=(14, 7), colormap='tab20')

# Add value labels
for c in ax.containers:
    labels = [f'{int(v.get_height())}' if v.get_height() > 0 else '' for v in c]
    ax.bar_label(c, labels=labels, label_type='center', fontsize=8)

plt.title("Churned Customers by Service Type (Stacked)")
plt.xlabel("Service Feature")
plt.ylabel("Number of Churned Customers")
plt.xticks(rotation=45)
plt.legend(title="Service Type")
plt.tight_layout()
plt.show()
```

```python


# List of service columns
service_columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Filter only churned customers
churned_df = df[df['Churn'] == 'Yes']

# Prepare data for plotting
plot_data = pd.DataFrame()

for col in service_columns:
    value_counts = churned_df[col].value_counts().reset_index()
    value_counts.columns = ['ServiceType', 'Count']
    value_counts['Service'] = col
    plot_data = pd.concat([plot_data, value_counts], ignore_index=True)

# Plotting
plt.figure(figsize=(16, 8))
ax = sns.barplot(
    data=plot_data,
    x='Service',
    y='Count',
    hue='ServiceType'
)

# Add value labels
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', fontsize=9)

plt.title("Churned Customers by Service Types")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python



# List of service columns
service_columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up the plot grid
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(20, 12))
axes = axes.flatten()  # to index subplots in a loop

# Plot countplots
for i, col in enumerate(service_columns):
    ax = sns.countplot(data=df, x=col, hue='Churn', ax=axes[i])
    ax.set_title(f'{col} vs Churn')
    ax.set_xlabel(col)
    ax.set_ylabel('Count')

    # Add labels to bars
    for container in ax.containers:
        ax.bar_label(container, fmt='%d', fontsize=9)

    ax.legend(title='Churn')

plt.tight_layout()
plt.show()
```
